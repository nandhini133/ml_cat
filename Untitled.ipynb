```python
import numpy as np
from matplotlib import pyplot as plt
import random
import pandas as pd
x = np.random.normal(2.0, 1.0, 1000)
y = (0.2*np.random.normal(8.0, 3.0, 1000))+x
plt.scatter(x,y)
```

```python
mx=sum(x)/len(x)
my=sum(y)/len(y)
print(mx,my)
```

```python
data=pd.DataFrame()
data['x']=x
data['y']=y
data
```

```python
data['x-x']=data['x']-mx
data['y-y']=data['y']-my
data['x-x2']=data['x-x']*data['x-x']
data['(x-x)(y-y)']=(data['y-y']*data['x-x'])
m=np.sum(data['(x-x)(y-y)'])/np.sum(data['x-x2'])
m
```

```python
c=my-m*mx
c
```

```python
x=data['x']
y=data['y']
pred_y=m*x+c
plt.scatter(x,y,s=7,c='black')
plt.plot(x,pred_y)
plt.show()
pred_y
```

```python
#using sklearn 
from sklearn.linear_model import LinearRegression
l=LinearRegression()
y = data['y'].values.reshape(-1, 1)
x = data['x'].values.reshape(-1, 1)
```

```python
l.fit(x, y)
```

```python
ms=l.intercept_
print(ms)
```

```python
cs=l.coef_
print(cs)
```

```python
print(m,l.coef_[0][0])
print(c,l.intercept_[0])
```
